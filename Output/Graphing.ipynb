```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
metrics = ["Accuracy", "Precision", "Recall", "F1-Score"]
models = ['Random Forest','SVM','Naive Bayes','AdaBoost','KNN']
datasets = ["non-ablated","spectral","spectral features and spectral indices","spectral features and textural features"]
```

```python
output_dir = '/content/drive/MyDrive/Images'
```

```python
na_train = pd.read_csv('/content/drive/MyDrive/ML results/Non-Ablated_Data/Chicago_2024_train.csv')
s_train = pd.read_csv('/content/drive/MyDrive/ML results/Spectral_features/Chicago_2024_train.csv')
st_train = pd.read_csv('/content/drive/MyDrive/ML results/Spectral+Textures/Chicago_2024_train.csv')
si_train = pd.read_csv('/content/drive/MyDrive/ML results/Spectral+Indices/Chicago_2024_train.csv')
```

```python
dfs_train = [na_train, s_train, st_train, si_train]
```

```python
na_test = pd.read_csv('/content/drive/MyDrive/ML results/Non-Ablated_Data/Chicago_2024_test.csv')
s_test = pd.read_csv('/content/drive/MyDrive/ML results/Spectral_features/Chicago_2024_test.csv')
st_test = pd.read_csv('/content/drive/MyDrive/ML results/Spectral+Textures/Chicago_2024_test.csv')
si_test = pd.read_csv('/content/drive/MyDrive/ML results/Spectral+Indices/Chicago_2024_test.csv')
```

```python
dfs_test = [na_test,s_test,st_test,si_test]
```

# Plotting

## Training

```python
os.path.join(output_dir, f"{metric}_comparison_training.png")
```

```python
for metric in metrics:
        plt.figure(figsize=(10, 6))
        x = np.arange(len(models))  # X positions for bars

        # Plot bars for single dataset
        plt.bar(x, na_train[metric], width=0.5, color='skyblue',label=metric)

        # Formatting
        plt.xticks(x, models, rotation=45)
        plt.xlabel("Model")
        plt.ylabel(metric)
        plt.title(f"Train data comparison for ML models")
        plt.legend()
        plt.grid(axis='y', linestyle='--', alpha=0.7)

        # Save plot
        save_path = os.path.join(output_dir, f"{metric}_train_single.png")
        plt.savefig(save_path, bbox_inches="tight", dpi=300)
        plt.close()  # Close plot to free memory
```

```python
# Set width of bars
bar_width = 0.2

# Generate plots for each metric
for metric in metrics:
    plt.figure(figsize=(10, 6))
    x = np.arange(len(models))  # X positions for bars

    # Plot bars for each file
    for i, df in enumerate(dfs_train):
        plt.bar(x + i * bar_width, df[metric], width=bar_width, label=datasets[i])

    # Formatting
    plt.xticks(x + bar_width * 1.5, models, rotation=20)
    plt.xlabel("Model")
    plt.ylabel(metric)
    plt.title(f"{metric} Comparison Across Models")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)



    # Save plot
    save_path = os.path.join(output_dir, f"{metric}_comparison_training.png")
    plt.savefig(save_path, bbox_inches="tight", dpi=300)
    plt.show()
    plt.close()  # Close plot to free memory
```
